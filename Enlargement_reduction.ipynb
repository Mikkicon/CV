```python
import cv2
import numpy as np
```

```python
img = cv2.imread("1.jpg")
```

```python
print(img.shape)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
plt.imshow(img)
```

```python
# 1-dimensional interpolation. Simply duplicating neighbor pixels
def enlargement(image):
    rows, cols = image.shape[:2]
    result = np.zeros((rows*2, cols*2, 3), np.uint8)
    for i in range(rows):
        for j in range(cols):
            kernel = image[i][j]
            result[i*2][j*2] = kernel
            result[i*2][(j*2)+1] = kernel
            result[(i*2)+1][j*2] = kernel
            result[(i*2)+1][(j*2)+1] = kernel
    cv2.imwrite("big.jpg", result)
    plt.imshow(result)
```

```python
# Removing every second pixel
def reduction(image):
    rows, cols = image.shape[:2]
    result = np.zeros((rows//2, cols//2, 3), np.uint8)
    for i in range(rows//2):
        for j in range(cols//2):
            kernel = image[i*2][j*2]
            result[i][j] = kernel
    cv2.imwrite("small.jpg", result)
    plt.imshow(result)
```

```python
enlargement(img)
```
